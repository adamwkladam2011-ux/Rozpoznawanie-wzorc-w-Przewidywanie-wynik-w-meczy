```python
# ============================================================
# ⚽ FOOTBALL-DATA.ORG  —  TOP 5 LIG EUROPY (sezony 2024 i 2025)
# Google Colab Ready ✅
# ============================================================

!pip install -q requests pandas

import requests
import pandas as pd
from datetime import datetime
from time import sleep

# === 🔑 Wklej swój TOKEN z https://www.football-data.org/ ===
API_TOKEN = "d7d427428ea141edbf4f88b2ed2e751c"

# === Ligi i sezony ===
LEAGUES = {
    "PL": "Premier_League",   # Anglia
    "FL1": "Ligue_1",         # Francja
    "BL1": "Bundesliga",      # Niemcy
    "PD": "La_Liga",          # Hiszpania
    "SA": "Serie_A"           # Włochy
}

SEASONS = [2024, 2025]        # pobieramy oba sezony
STATUS = "FINISHED"           # lub "SCHEDULED", "LIVE"

headers = {"X-Auth-Token": API_TOKEN}
all_matches = []

# === Pobieranie danych dla każdej ligi i sezonu ===
for season in SEASONS:
    for code, name in LEAGUES.items():
        print(f"\n⏳ Pobieram {name} ({code}), sezon {season}/{season+1} ...")
        url = f"https://api.football-data.org/v4/competitions/{code}/matches?season={season}&status={STATUS}"

        try:
            response = requests.get(url, headers=headers)
            if response.status_code != 200:
                print(f"❌ Błąd {response.status_code}: {response.text}")
                continue

            data = response.json()
            matches = data.get("matches", [])
            print(f"✅ {name} {season}: {len(matches)} meczów")

            df = pd.DataFrame([
                {
                    "liga": name,
                    "kod_ligi": code,
                    "id_meczu": m["id"],
                    "data": datetime.fromisoformat(m["utcDate"].replace("Z", "+00:00")).strftime("%Y-%m-%d %H:%M"),
                    "gospodarz": m["homeTeam"]["name"],
                    "goście": m["awayTeam"]["name"],
                    "wynik": (
                        f"{m['score']['fullTime']['home']} - {m['score']['fullTime']['away']}"
                        if m['score']['fullTime']['home'] is not None else "-"
                    ),
                    "status": m["status"],
                    "kolejka": m.get("matchday", None),
                    "sezon": season
                }
                for m in matches
            ])

            # zapis do osobnego CSV
            filename = f"{name}_{season}.csv"
            df.to_csv(filename, index=False, encoding="utf-8-sig")
            print(f"💾 Zapisano {filename}")

            all_matches.append(df)

            # mała pauza, żeby nie przekroczyć limitów API
            sleep(1.5)

        except Exception as e:
            print(f"⚠️ Błąd przy {name} {season}: {e}")

# === Scalanie wszystkiego w jeden plik ===
if all_matches:
    df_all = pd.concat(all_matches, ignore_index=True)
    df_all.to_csv("all_matches.csv", index=False, encoding="utf-8-sig")
    print("\n🏁 Zapisano łączny plik: all_matches.csv")
    print(f"🔢 Łącznie {len(df_all)} meczów z {len(SEASONS)} sezonów i {len(LEAGUES)} lig.")
    display(df_all.sample(15))
else:
    print("⚠️ Nie udało się pobrać żadnych danych.")
```
